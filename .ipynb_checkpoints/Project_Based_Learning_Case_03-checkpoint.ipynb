```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import joblib
```

```python
#Load Dataset and show total missing value
movies = pd.read_csv('imdb_top_2000_movies.csv')
movies
print("Missing Value before drop Na")
print(movies.isnull().sum()) #--> Cek missing volue each coloumn
print(movies.shape)
movies = movies.dropna(axis=0)
print("")
print("Missing Value after drop Na")
print(movies.isnull().sum())
print(movies.shape)
```

```python
#Cek duplicate film 
movies[movies[['Movie Name','Release Year','Director']].duplicated()] #no duplicate
```

```python
movies.dtypes #check variabel metascore and IMDB rating
```

```python
#Mencari hubungan antara variabel metascore and rating -->linier
plt.figure(figsize=(15,10))
sns.scatterplot(data=movies, y='IMDB Rating',x='Metascore')
plt.title('Correlation Metascore and IMDB Rating',fontsize=25)
```

```python
#Mencari nilai korelasi antar variabel
movies['Metascore'].corr(movies['IMDB Rating']) 
```

```python
#Deklarasi target dan input/feature
X = movies['Metascore']
Y=  movies['IMDB Rating']
```

```python
#Cek jumlah data target dan feature
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,train_size=0.75,random_state=0000)
print(f'X Train = {X_train.shape}')
print(f'X Test = {X_test.shape}')
print(f'Y Train = {Y_train.shape}')
print(f'Y Test = {Y_test.shape}')
```

```python
X_train=X_train.to_frame()
Y_train=Y_train.to_frame()
```

```python
#Proses Learning
lr= LinearRegression()
lr.fit(X_train, Y_train)
```

```python
#Mencari nilai intercept dan koefisien
print(f'Intercept = {lr.intercept_}')
print(f'Coefission = {lr.coef_}')
```

```python
# Persamaan:   Y= 0.0375X+4.633
```

```python
#Proses Test
X_test=X_test.to_frame()
Y_predict= lr.predict(X_test)
```

```python
plt.figure(figsize=(15,10))
plt.plot(range(1,469), Y_predict,color='red')
plt.plot(range(1,469), Y_test,color='blue')
plt.xlabel('Index',fontsize=20)
plt.ylabel('Imdb Rating',fontsize=20)
plt.title('Hasil prediksi vs Test IMDB Rating',fontsize=25)
```

```python
plt.figure(figsize=(15,10))
plt.plot(range(1,469), Y_predict.flatten()-Y_test,color='purple')
plt.xlabel('Index',fontsize=20)
plt.ylabel('Imdb Rating',fontsize=20)
plt.title('Selisih Error Hasil prediksi vs Test IMDB Rating',fontsize=25)
```

```python
#Menghitung nilai MSE
mean_squared_error(Y_test,Y_predict)
```

```python
#Menghitung nilai R_Square
r_squared= r2_score(Y_test,Y_predict)
r_squared
```

```python
#Hubungan Y_test and Y_Predict
plt.scatter(Y_test,Y_predict)
plt.title('Hubungan antara Y_Test dan Y_predict')
```

```python
#Menyimpal model ML
joblib.dump(lr, 'model_regresi_linear.pkl')
```

```python
#Load ML untuk dicoba
lr_loaded = joblib.load('model_regresi_linear.pkl')
```

```python
#Mencoba ML untuk menebak score
metascore = pd.DataFrame({'Metascore':[100]})
predicted_imdb_score = lr_loaded.predict(metascore)
print(f'Prediksi skor IMDb dengan metascore {metascore.values} adalah: {predicted_imdb_score}')
```

Recommander System :
Merekomendasikan Film Box office dengan IMDB rating minimal 6 dan dengan vote minimal 100 ribu

```python
movies['Votes']= movies['Votes'].str.replace(',','').astype(int) # mengubah datatype votes dari str ke int
```

```python
# popular_movie= movies[(movies['IMDB Rating']>6) & (movies['Votes']>350000)][['Movie Name','Votes','IMDB Rating']]
```

```python
movies_filter = movies[['Movie Name','Votes','IMDB Rating']]
```

```python
def movie_recommend(k,votes,imdb):
    popular_movies = movies_filter[(movies_filter['IMDB Rating']>imdb) & (movies_filter['Votes']>votes)].sort_values(by=['IMDB Rating','Votes'], ascending=False)
    return popular_movies['Movie Name'].head(k)
```

```python
movie_recommend(10,votes=10000,imdb=6.7) # Contoh penggunaaan Fungsi : mencari 10 film dengan vote minimal 10 rb dengan rating >6.7 
```

```python
from surprise import KNNBasic, Dataset, Reader, accuracy
from surprise.model_selection import train_test_split, GridSearchCV
```

```python
movies2= movies[['Movie Name','IMDB Rating','Votes']]
```

```python
movies2
```

```python
movies2.info()
```

```python
reader = Reader(rating_scale=(0,5))
data= Dataset.load_from_df(movies2,reader)
trainset, testset = train_test_split(data, test_size=0.2,random_state=42)
```

```python
parameter_grid = {
    'k':[20,30,40],
    'min_k':[3,6,9],
    'sim_options':{'name':['msd','pearson','cosine'],
                   'user_based':[True]}}        

gsvc= GridSearchCV(KNNBasic, parameter_grid,cv=4, measures=['rmse'])
```

```python
gsvc.fit(data)
```

```python
gsvc.best_params
```

```python
gsvc.best_score
```

```python
knn_user= KNNBasic(**gsvc.best_params)
knn_user.fit(trainset)
```

```python
predict= knn_user.test(testset)
accuracy.rmse(predict)
```
